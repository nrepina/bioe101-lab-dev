# EKG Signal Processing #

This demo will show that we can use digital filtering to clean up a raw signal and also use cross-correlation to find where the peaks of the QRS complexes are of the signal. We can use this to determine if the heart rate is regular and normal.

First we import the necessary packages

```python
%matplotlib inline
import serial
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft
```

### Data Acquisition ###

I've set up a circuit that collects the ekg data. I've set the microcontroller to sample at a rate of ~250 samples per second, which is more than twice the normal bandwidth of an EKG signal. I've already performed lowpass filtering at half the sampling frequency to prevent aliasing as well.

```python
device = "COM8"
baud = 115200
ser = serial.Serial(device, baud, timeout=10)
```

```python
raw_data = []
while (len(raw_data) < 2000):
    data = float(ser.readline())
    raw_data.append(data)
ser.close()

```

### Digital Processing on the Raw Data ###

We plot the data (zoomed out and in) and also calculate the DFT of it.

```python
with open("raw_data.txt", "w") as data:
    for i in raw_data:
        data.writelines(str(i)+'\n')
```

```python
# ser.close()
# signal = raw_data
with open("raw_data_old.txt", "r") as raw_data_txt:
    loaded_data = raw_data_txt.readlines()
signal = loaded_data
dft = fft(signal)
plt.plot(signal[0:len(signal)])
plt.figure()
plt.plot(signal[:100])
```

There is a noticable interference in the signal due to powerline interference. The human body is capacitively coupled with the AC power in the wall, so it shows up in the signal as well. Even with a good analog notch filter, we can't get rid of most of the 60 Hz interference without losing some of our signal.



### DFT Plotting and Filtering ###

```python
fs = 2000
N = len(dft)
freqs = [i*fs/N for i in range(N)]
plt.plot(freqs[50:600], np.abs(dft[50:600]))
```

Here we perform a bandpass filter combined with a notch filter to remove the 60 Hz.

$$H_{hp} = \frac{s^2}{s^2 + w_h s + w_h^2}$$
$$H_{lp} = \frac{1}{s^2 + 2w_p s + w_p^2}$$
$$H_{n} = \frac{s^2 + w_n^2}{s^2 + w_n s + w_n^2}$$

$$w_n = 2\pi 60, w_h = 2\pi 100, w_l = 2\pi 0.5$$

```python
#### fitering ###

def notch_filter(freq, energy):
    w = 2*np.pi*freq
    s = complex(0,1)*w
    wn = 2*np.pi*60
    wl = 2*np.pi*100
    wh = 2*np.pi*0.5
    notch = (s**2 + (wn)**2)/(s**2 + wn*s + wn**2)
    high_pass = (s**2)/(s**2 + wh*s + wh**2)
    low_pass = 4000000/(s**2 + 2*wl*s + wl**2)
    return energy*notch*low_pass*high_pass

dft_filtered = [notch_filter(freqs[i], dft[i]) for i in range(N)]
plt.plot(freqs[0:700], np.abs(dft_filtered[0:700]))
```

```python
#### reconstruction ####
filtered_signal = ifft(dft_filtered)
plt.plot(signal[:500])
plt.figure()
plt.plot(filtered_signal[:500])
```

```python
signal_period = filtered_signal[80:180]
corrs = np.correlate(signal_period,filtered_signal)
plt.plot(corrs)
```

### Peak detection ###
Peaks seem to happen when the cross correlation has a value greater than ~400000, so that'll be the threshold.

```python
thresh = 1000000
peak_times = [corrs[i] if corrs[i] > thresh else 0 for i in range(len(corrs))]
plt.plot(peak_times)
```

### Naive Beat Detection Algorithm ###
Using the cross correlation data I found above, I then wrote a ghetto beat-detection algorithm to get all the peaks of the QRS complex to reliably get the averaged heartrate and then classify the heartrate

```python
def heart_rate(ekg_peaks):
    time_between_peaks = []
    curr_time = 0
    i = 0
    while ekg_peaks[i] == 0:
        i +=1
    while i < len(ekg_peaks):
        if ekg_peaks[i] == 0:
            curr_time += 1
        else:
            if curr_time != 0:
                time_between_peaks.append(curr_time)
            curr_time = 0
        i += 1
    return 1.5*np.average(time_between_peaks)/fs * 60 

heart_rate = heart_rate(peak_times)

def heart_rate_classify(heart_rate):
    if heart_rate > 100:
        return str(heart_rate) + " bpm is tachychardic"
    elif 50 <= heart_rate <= 100:
        return str(heart_rate) + " bpm is normal"
    elif heart_rate < 50:
        return str(heart_rate) + " bpm is bradychardic"
    
condition = heart_rate_classify(heart_rate)
print(condition)
```
